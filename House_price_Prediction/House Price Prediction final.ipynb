Since we are going to make the prediction so I am going to use the regression algorithms. Before I start, I would like to set my objective like, I am going to use the linear regression technique and Elastic net regression, and I will also be using Random forest regression and Gradient boosting algorithm and compare with the polynomila regression technique and elastic technique. Finally I will be using the Neural network for the prediction.

I am setting the objective because every algorithm have there own assumptions and every algorithm required different kind of data so we need to perform the data preprocessing steps particularlly for those algorithms

For the linear regression, it has 4 basic assumptions, they are:-
    - Linearity: The relationship between X and the mean of Y is linear.
    - Homoscedasticity: The variance of residual is the same for any value of X.
    - Independence: Observations are independent of each other.
    - Normality: For any fixed value of X, Y is normally distributed.
> Therefore, we need to check all this before we implement the linear regression model for the better result.

#### ElasticNet Regression:-
> the elastic net is a regularized regression method that linearly combines the L1 and L2 penalties of the lasso and ridge methods.(it overcomes the limitation of lasso and the ridge methods)
    - Ridge regression(L2 regularization) tends to shrink all the correlated variable together(when most of the features are usefull then we use ridge regression technique)
    -  Lasso regression(L! regularization) tends to peak on of the correlated feature and remove other correlated variable. it tries to make the model simple and best removing the correlated variables(when we have alot of features that correlated and may not be usefull then we we lasso regression). Therefore, Lasso regression can also  be used for the feature selection.
   
- When there are a tons of features and we will be loosing a lot of time going through each and every variable individually so we need to use some technique so that we can get better result even without taking much of a care of those variables my ourself. One of the technique is elasticnet regression.
- elasticNet is therefore usefull when there exist a correlation between the variable and we have alot of variables 
- for the hyper parameter tuning we need to play with the lamda value.(0 to any positive value)

#### Random Forest Regression:
> Fully grown Decision tree leads to the Low Bias and High Bias so we need to prun the tree inorder to make low bias and low variance or simple to remove the overfitting. Hence in the random forest what is done is that, sample of row and column will be taken(with replacement) and constructd a base model named as m1 and similarly many trees will be constructed with different sample(also known as bootstraping), finally majority vote will be counted and the final prediction will be made(In the case of classification) where as in the regression, mean or median of all the model will be taken depending upon the distribution of the data. Due to that result will not be only dependent on the one decision tree wich leads to the low variance.(For the hyper parameter tuning we need to play with the number of trees)
- Note:- since in the random forest, it will be taking the random sample of the data so when ever the small portion of the data is changed or the test data is changed then it will not be the problem for the random forest it can handle those situation very well.

#### Gradient Boosting algorithm: 
> This is an ensemble technique(also known as boosting techniques) where it uses decision tree as a base model and using the sequential decision trees for updating the model to reduce the error  and predict the final value. I will be using this algorithm also to predict the house price and we can see how good the ensemble techniques are.

Finally we can compare all the model with the Neural Network, as many of us think that Artificial Neural Network perform always better. So I will be building ANN finally to see which model will be the best.

```python
# defining the function to calculate the performance metrics
def performance_metrixs(regressor, actual, predicted):
    print('Performance Metrics of the model :', regressor )
    print('mean_absolute_error : ', mean_absolute_error(actual,predicted))
    print('mean_squared_error : ', mean_squared_error(actual,predicted, squared = False))
    print('R_squared : ', r2_score(actual,predicted))
    print('-----------------------------------------------\n')
    
def plotting_result(actual,predicted):
    x_axis = range(len(actual))
    plt.scatter(x_axis, actual, color = 'blue')
    plt.plot(x_axis, predicted, color = 'red')
    plt.legend(['predicted','actual'])
    plt.ylabel('SalePrice')
    plt.show()
```

```python
%%time
def prediction():
    for i,(regressor_name,regressor) in enumerate(regressor_dict.items()):
        if regressor_name == 'linear_regression':
            regressor.fit(X_train,y_train)
            y_pred = regressor.predict(X_test)
            print('Coefficient : {}'.format(regressor.coef_))
            print('Intercept : {}'.format(regressor.intercept_))
            performance_metrixs(regressor_name,y_test,y_pred)
            plotting_result(y_test,y_pred)
            print('##########################################################')
        elif regressor_name == 'elastic_net':
            print(regressor)
            for hyperpara in parameters.items():
                if hyperpara[0] == regressor_name:
                    hyperparameter = hyperpara[1]
            rscv = RandomizedSearchCV(regressor, param_distributions=hyperparameter, cv = 5, n_jobs=-1, n_iter=20) 
            result = rscv.fit(X_train,y_train)
            y_pred = rscv.predict(X_test)
            performance_metrixs(regressor_name,y_test,y_pred)
            plotting_result(y_test,y_pred)
            print(f'Best parameters are: {result.best_params_}')
            print('##########################################################')
        elif regressor_name == 'svr':
            print(regressor)
            for hyperpara in parameters.items():
                if hyperpara[0] == regressor_name:
                    hyperparameter = hyperpara[1]
            rscv = RandomizedSearchCV(regressor, param_distributions=hyperparameter, cv = 5, n_jobs=-1, n_iter=20) 
            result = rscv.fit(X_train,y_train)
            y_pred = rscv.predict(X_test)
            performance_metrixs(regressor_name,y_test,y_pred)
            plotting_result(y_test,y_pred)
            print(f'Best parameters are: {result.best_params_}')
            print('##########################################################')
        elif regressor_name == 'random_forest_regressor':
            print(regressor)
            for hyperpara in parameters.items():
                if hyperpara[0] == regressor_name:
                    hyperparameter = hyperpara[1]
            rscv = RandomizedSearchCV(regressor, param_distributions=hyperparameter, cv = 5, n_jobs=-1, n_iter=20) 
            result = rscv.fit(X_train,y_train)
            y_pred = rscv.predict(X_test)
            performance_metrixs(regressor_name,y_test,y_pred)
            plotting_result(y_test,y_pred)
            print(f'Best parameters are: {result.best_params_}')
            print('##########################################################')
        else:
            print(regressor)
            for hyperpara in parameters.items():
                if hyperpara[0] == regressor_name:
                    hyperparameter = hyperpara[1]
            rscv = RandomizedSearchCV(regressor, param_distributions=hyperparameter, cv = 5, n_jobs=-1, n_iter=20) 
            result = rscv.fit(X_train,y_train)
            y_pred = rscv.predict(X_test)
            performance_metrixs(regressor_name,y_test,y_pred)
            plotting_result(y_test,y_pred)
            print(f'Best parameters are: {result.best_params_}')
            print('##########################################################')
    return
```

```python
ann_model = Sequential()
ann_model.add(Dense(units = 79, activation = 'relu' , input_dim = 79, kernel_initializer = 'normal'))
ann_model.add(tensorflow.keras.layers.Dropout(0.2))
ann_model.add(Dense(units = 160, activation = 'relu'))
ann_model.add(tensorflow.keras.layers.Dropout(0.2))
ann_model.add(Dense(units = 100, activation = 'relu'))
ann_model.add(tensorflow.keras.layers.Dropout(0.2))
ann_model.add(Dense(units = 50, activation = 'relu'))
ann_model.add(tensorflow.keras.layers.Dropout(0.2))
ann_model.add(Dense(units = 10, activation = 'relu'))
ann_model.add(tensorflow.keras.layers.Dropout(0.2))
ann_model.add(Dense(units = 1, activation = 'linear'))
```

```python
%%time
ann_model.compile(loss = 'mse', optimizer = 'adam', metrics = ['mse','mae'])
history = ann_model.fit(X_train,y_train, epochs = 300, batch_size= 10, validation_split= 0.3 )
y_pred = ann_model.predict(X_test)
```
